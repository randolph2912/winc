```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample dataframe
df = pd.DataFrame({'Country': ['USA', 'China', 'Germany', 'UK', 'Japan'],
                   'GDP': [21000, 14000, 5000, 6000, 5000]})

# Plot the data
df.plot(kind='bar', x='Country', y='GDP', color='green')

# Add a title to the plot
plt.title('GDP by Country')

# Show the plot
plt.show()

%matplotlib inline
```
