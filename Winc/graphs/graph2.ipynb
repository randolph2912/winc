```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Country': ['United States', 'China', 'India', 'Russia', 'Japan'],
        'Rel_1990': [1, 2, 3, 4, 5],
        'Rel_2005': [3, 4, 2, 1, 5],
        'Rel_2017': [2, 1, 3, 5, 4]}

df = pd.DataFrame(data)

# sort the dataframe based on the values of Rel_2017 in descending order
df.sort_values(by='Rel_2017', ascending=False, inplace=True)

# select the top three countries
top_three = df.head(3)

years = [1990, 2005, 2017]
fig, ax = plt.subplots()
for i in range(3):
    ax.plot(years, [top_three.iloc[i]['Rel_1990'], 
                    top_three.iloc[i]['Rel_2005'], 
                    top_three.iloc[i]['Rel_2017']], 
            label=top_three.iloc[i]['Country'])

ax.set_xlabel('Year')
ax.set_ylabel('Relative Value')
ax.legend()
ax.set_title('Top Three Countries by Relative Value')

plt.show()
```
