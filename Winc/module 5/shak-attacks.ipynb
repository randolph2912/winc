```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
url = "https://raw.githubusercontent.com/ibmw/Shark-Attack/master/shark_attack.csv"
sharks = pd.read_csv(url, encoding='ISO-8859-1')
```

```python
# Count the number of attacks by activity
attacks_by_activity = sharks.groupby("Activity").size().reset_index(name="Attack Count")
```

```python
# Sort the activity by number of attacks
attacks_by_activity = attacks_by_activity.sort_values("Attack Count", ascending=False)
```

```python
# Plot the number of attacks by activity
attacks_by_activity.plot(kind="bar", x="Activity", y="Attack Count", legend=False)
plt.xlabel("Activity")
plt.ylabel("Attack Count")
plt.title("Number of Attacks by Activity")
plt.tight_layout()  # Add this line to adjust subplot parameters
plt.show()
```
